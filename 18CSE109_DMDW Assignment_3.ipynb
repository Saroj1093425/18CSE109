```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df1 =pd.read_csv("D:\\DATASET\\googleplaystore.csv",nrows =1000)
df1.shape
```

```python
df1.head()
```

```python
x = df1["Rating"]
y = df1["Reviews"] 
plt.scatter(x,y)
plt.title("Google play store")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.plot(x,y,"g^--",linewidth =3,markersize = 10,label = "Rating vs Review")
plt.title("Rating-Review",fontsize = 30)
plt.xlabel("Rating",fontsize =20)
plt.ylabel("Review",fontsize =20)
plt.legend(loc = 4)
plt.show()
```

```python
plt.hist(df1["Rating"])
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(y,bins,rwidth = 0.8,histtype = "bar")
plt.title("Play store")
plt.xlabel("Bins")
plt.ylabel("Review")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c= "r",marker = "*",s = 500,alpha =0.2,linewidth=5, edgecolor="g")                   # verts = "<"
plt.title("Google play store")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c= "r",marker = "*",s = 500,alpha =0.2,linewidth=5, edgecolor="g") 
plt.scatter(x,df1["Installs"],c= "y",marker = "o",s = 500,alpha =0.2,linewidth=5, edgecolor="g") # verts = "<"
plt.title("Google play store")
plt.xlabel("Rating")
plt.ylabel("Reviews and Install")
plt.show()
```
